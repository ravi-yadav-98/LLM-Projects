```python
client = openai.OpenAI(
    api_key="ollama",  # dummy value, Ollama ignores it
    base_url="http://localhost:11434/v1"
)

def get_response(prompt, model = 'gemma3:1b'):
    completion = client.chat.completions.create(
    model=model,
    messages=[
        {"role": "user", "content": prompt},
    ],
    )
    
    response = completion.choices[0].message
    return response.content
```

#### Experimental Setup

In the multi-document question answering task, the model inputs are:

- (i) a question to answer, and  
- (ii) *k* documents (e.g., passages from Wikipedia),

where exactly **one** document contains the answer to the question, and the remaining *k − 1* documents are **distractors** that do not contain the answer.

**This task requires the model to access the document that contains the answer within its input context and use it to answer the question**
